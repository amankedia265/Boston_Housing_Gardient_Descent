```python
import numpy as np
import pandas as pd
import sklearn as skt
```

```python
data = np.loadtxt("train.csv",delimiter=",")
print(data)
```

```python
def step_gradient(points, learning_rate, m):
        M, N = data.shape
        m_slope = np.zeros(N)
        for j in range(M):
            y = points[j,-1]
            x = points[j,:]
            x[-1] = 1
            for i in range(N):
                m_slope[i] += ((-2/M)*(y-(m*x).sum())*x[i])
        new_m = m-learning_rate*m_slope
        return new_m

def cost(points, m):
    total_cost = 0
    M, N = data.shape
    for i in range(M):
            y = points[i,-1]
            x = points[i,:]
            x[-1]=1
            total_cost += (1/M)*((y - (m*x).sum())**2)
    return total_cost

def gd(points, learning_rate,num_iterations):
    M, N = data.shape                         
    m = np.zeros(N)
    for i in range(num_iterations):
        m = step_gradient(points, learning_rate, m)
        print(i,"Cost:", cost(points,m))
    return m
def run():
    learning_rate = 0.1
    num_iterations = 500
    m = gd(data, learning_rate, num_iterations)
    print(m)
    
run()
```
